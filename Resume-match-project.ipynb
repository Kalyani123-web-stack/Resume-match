```python
# Extract Text from the Uploaded Resume

import fitz  # PyMuPDF for PDF text extraction

def extract_text_from_pdf(pdf_path):
    """Extracts text from a PDF resume."""
    doc = fitz.open(pdf_path)
    text = "\n".join([page.get_text() for page in doc])
    return text

# Use your uploaded file
pdf_path = "sample_resume.pdf"
resume_text = extract_text_from_pdf(pdf_path)

print("Extracted Resume Text:\n", resume_text)
```

```python
# Extract Skills and Experience

import re
import string
import nltk
from nltk.corpus import stopwords

# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Function to extract skills
def extract_skills(text):
    skills_list = {"python", "django", "flask", "javascript", "sql",
                   "machine learning", "tensorflow", "aws", "git", "docker",
                   "pandas", "numpy", "nlp", "scikit-learn", "rest api",
                   "mysql", "postgresql"}

    text = text.lower().translate(str.maketrans('', '', string.punctuation))
    words = text.split()
    words = [word for word in words if word not in stop_words]
    text_cleaned = " ".join(words)

    found_skills = {skill for skill in skills_list if re.search(rf'\b{re.escape(skill)}\b', text_cleaned)}
    return list(found_skills)

# Function to extract experience
def extract_experience(text):
    experience_pattern = r"(\d+)\s*(?:years?|yrs?\b|\+)"
    matches = re.findall(experience_pattern, text)
    return max(map(int, matches), default=0)

# Extract skills and experience from resume text
skills = extract_skills(resume_text)
experience = extract_experience(resume_text)

# Print results
print("\nExtracted Skills:", skills)
print("Years of Experience:", experience)
```

```python
resume_skills = ['python', 'django', 'flask', 'javascript', 'sql',
                 'machine learning', 'tensorflow', 'aws', 'git', 'docker']
resume_experience = 5  # Extracted from your resume
```

```python
job_skills = ['python', 'django', 'flask', 'machine learning', 'tensorflow', 'aws']
job_experience = 3  # Job requires 3 years of experience
```

```python
def match_resume(resume_skills, job_skills, resume_experience, job_experience):

    skill_match = len(set(resume_skills) & set(job_skills)) / len(job_skills) * 100  # Skill match %-Finds common skills between resume and job.

    experience_match = 100 if resume_experience >= job_experience else (resume_experience / job_experience) * 100  # Experience %(5>3,so 100%) (if resume_exp 2< job_exp 4 then:(2/4)*100=50%)

    overall_score = (0.7 * skill_match) + (0.3 * experience_match)  # Weighted Score (70% weight is given to skills.30% weight is given to experience.)

    return round(overall_score, 2),skill_match, experience_match

match_score, skill_match, experience_match = match_resume(resume_skills, job_skills, resume_experience, job_experience)
print(f"Skill Match: {len(set(resume_skills) & set(job_skills))}/{len(job_skills)} ({skill_match}%)")
print(f"Experience Match: {resume_experience}/{job_experience} years ({experience_match}%)")
print(f"Overall Resume Match Score: {match_score}%")
```
